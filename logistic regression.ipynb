```python
import pandas as p 
import numpy as n
import matplotlib.pyplot as m
from sklearn.linear_model import LinearRegression  as l
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
data=p.read_csv('diabetes.csv',header=None,names=col_names)
d=data[1:]
d
```

```python
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
x=d[feature_cols]
y=d.label
from sklearn.model_selection import train_test_split as t
x_train,x_test,y_train,y_test=t(x,y,test_size=0.20,random_state=42)
print(len(x_train))
print(len(x_test))
```

```python
obj=l()
obj=obj.fit(x_train,y_train)
y_pred=obj.predict(x_test)
from sklearn.metrics import classification_report as cr ,confusion_matrix as cm ,accuracy_score as a
result=cr(y_test,y_pred)
print(result)
```
