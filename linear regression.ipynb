```python
import numpy as n
import matplotlib.pyplot as m
import pandas as p 
import seaborn as s 
from  sklearn.model_selection import train_test_split as t
from sklearn.linear_model import LinearRegression as l
data=p.read_csv('weather.csv')
d=p.DataFrame(data)
d.head(10)
```

```python
m.subplot(221)
s.distplot(d['MaxTemp'])
m.subplot(222)
s.distplot(d['MinTemp'])
```

```python
x= d['MinTemp'].values.reshape(-1,1)
print(len(x))
y=d['MaxTemp'].values.reshape(-1,1)
print(y)
```

```python
x_train,x_test,y_train,y_test=t(x,y,test_size=0.2,random_state=0)
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
obj=l()
obj.fit(x_train,y_train)
a=obj.score(x_train,y_train)
b=obj.score(x_test,y_test)
print('{:.2f}'.format(a))
print('{:.2f}'.format(b))
```
