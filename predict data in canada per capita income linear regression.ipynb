```python
import numpy as n
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
d=pd.read_csv('canada1.csv')
d
```

```python
plt.scatter(d.year,d.percapitaincome,color='red',marker='o')
plt.xlabel('year')
plt.ylabel('percapitaincome')
plt.show()
```

```python
obj=linear_model.LinearRegression()
obj.fit(d[['year']],d.percapitaincome)
```
